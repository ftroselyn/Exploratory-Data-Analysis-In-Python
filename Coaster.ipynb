# Data Cleaning


```python
# Subsetting a list of Column(Removing columns that are not needed)


df1= df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
       #'Type', 
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
      # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
      # 'Track layout', 'Fastrack available', 'Soft opening date.1',
      # 'Closing date', 
    #'Opened',
    #'Replaced by', 'Website',
      # 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced',
         'latitude', 'longitude', 
         'Type_Main',
       'opening_date_clean', 
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
    #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']]
```

```python
df1['opening_date_clean'] = pd.to_datetime(df1['opening_date_clean'])
```

```python
# Renaming the columns in the df1

df1.columns
```

```python
# Renaming the columns in the df1
df2 = df1.rename(columns={'coaster_name':'C_Name',
            'year_introduced':'Year_into',
           'opening_date_clean': 'Opening_date',
            'speed_mph':'Speed_mph',
            'height_ft':'Height_ft',
           'Inversions_clean':'Inversions',
           'Gforce_clean':'Gforce'})
```

```python
df2
```

```python
# Check for missing values
missing_values = df2.isnull().sum()

print("Missing values per column:")
print(missing_values)
```

```python
# Check for duplicates using .loc
duplicate_rows = df2[df2.duplicated()]
print("Duplicate rows:")
print(duplicate_rows)
```

```python
df2.loc[df2.duplicated(subset=['C_Name'])].head(5)
```

```python
# Let check why We have duplicated rows, do a query command
# Checking an ec=xample duplicate


df2.query('C_Name  == "Crystal Beach Cyclone"')
```

```python
df2.duplicated(subset=['C_Name', 'Location', 'Opening_date'])
```

```python
# We going to run a reset, so our index does not jump up when dropping duplicates 

df2.loc[~df2.duplicated(subset=['C_Name', 'Location', 'Opening_date'])].reset_index()
```

```python
dfNew = df2.loc[~df2.duplicated(subset=['C_Name', 'Location', 'Opening_date'])].reset_index(drop=True)
```

# Ask a question about the data

```python
# Using plot or statistics
# What is the location with the fastest roller coasters(minimum of ten)?

dfNew['Location'].value_counts()
```

```python
#filter out where the Location is not equal

ax = dfNew.query('Location !="other"').groupby('Location')['Speed_mph'].agg(['mean', 'count']).query('count >= 10').sort_values('mean')['mean'].plot(kind='barh', figsize=(12, 5), title= 'Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
